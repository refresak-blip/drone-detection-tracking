This notebook describes the training process for the drone detection model.

The goal is to train YOLO models to detect drones in images and videos.

Two YOLO models were trained:

- YOLO11n as a baseline model
- YOLO11s as an improved model

The improved model was selected because it achieved better detection metrics.

```python
## Results

The improved YOLO11s model achieved the best performance.

| Model | Precision | Recall | mAP50 | mAP50-95 |
|------|-----------|--------|-------|----------|
| YOLO11n | ~0.954 | ~0.914 | ~0.954 | ~0.776 |
| YOLO11s | 0.975 | 0.950 | 0.972 | 0.813 |

YOLO11s was selected as the final model for detection and tracking.
```

## Conclusion

YOLO11s achieved better results than YOLO11n and was used as the final detection model.

The trained model was later combined with ByteTrack for video tracking.
